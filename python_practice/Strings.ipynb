## Interpolating Variables Into a String
In Python version 3.6, a new string formatting mechanism was introduced. This feature is formally named the Formatted String Literal, but is more usually referred to by its nickname <b>f-string</b>.


```python
n = 20
m = 25
prod = m * n
print('The product of',n,'and',m,'is',prod)
```

```python
n = 20
m = 25
prod = m * n
print('The product of {n} and {m} is {prod}')
print(f'The product of {n} and {m} is {prod}')
```
